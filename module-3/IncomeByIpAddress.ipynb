# Adds a new section about median income
We are using an external lib called `uszipcode` that has many features including getting the median income based in zip code.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

housing = pd.read_csv('https://raw.githubusercontent.com/wmezadev/CSE-450-TEAM-4/elias-income/incomes.csv')

y = housing['price']
x = housing['median_household_income']

plt.figure(dpi=100, figsize=(14, 4))
plt.xlabel('Income', fontsize=10)
plt.ylabel('Price (in million)', fontsize=10)

plt.title("Relationship between income and price", fontsize="small")
plt.scatter(x.head(50), y.head(50))

z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),"r--")

plt.show()
```
